```python
from bot.services.config_service import ConfigService
from pymongo import MongoClient
from bson import Int64
import datetime

config = ConfigService("config/config.yaml").load()

collection = MongoClient(config.mongoUri)["DiscordScrapeBot"]["Messages"]

guild_id = 1194024443362680874

time_threshold = (datetime.datetime.now(datetime.timezone.utc) - datetime.timedelta(minutes=30)).replace(tzinfo=None)

doc = list(collection.find({"guild_id": Int64(1194024443362680874), "timestamp": {"$gte": time_threshold}}).sort({"timestamp": 1}).limit(5))
doc
```

```python
messages = list(collection.find({"guild_id":  Int64(guild_id), "timestamp": {"$gte": time_threshold}}).sort({"timestamp": 1}).limit(5))
len(messages)
```
